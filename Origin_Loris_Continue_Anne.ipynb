### Afficher les airports sur une carte 

```python
import numpy
import pandas as pd
import flightradar24

fr = flightradar24.Api()
```

```python
airports_dict = fr.get_airports(); 
airports = pd.DataFrame.from_dict(airports_dict)

airports_details_serie = airports['rows']
type(airports_details_serie) # pandas.code.series.Series
airport_details = airports_details_serie.to_frame();
airports_details_serie.iloc[1]
    
```

```python
value_list_dict = list(airports.values)[1][1].keys()
list_valeurs = pd.DataFrame.from_dict(value_list_dict)
list_valeurs
```

```python
list_airport = []
for value in airports.values:
    list_airport.append(dict(zip(list_valeurs, value)))
    
df_airports = pd.DataFrame(list_airport, index = ["Airport "+ str(i) for i in range(len(list_airport))])
df_airports           
```

```python
value_list_dict = airports.values#[3][1]['name']
rows_a = []

for i in range(0,len(airports.values)):
    rows_a.append([value_list_dict[i][1]['name'], value_list_dict[i][1]['iata'], value_list_dict[i][1]['icao'], value_list_dict[i][1]['lat'], value_list_dict[i][1]['lon'], value_list_dict[i][1]['country'], value_list_dict[i][1]['alt']])

airports_df = pd.DataFrame(rows_a, columns=['name', 'iata', 'icao', 'lat', 'lon', 'country', 'alt'])
airports_df
```

```python
import folium as f
```

```python
map_airports = f.Map(
    location = [45.7640, 4.8357],
    zoom_start = 5
)
airports_df_petit_base = airports_df.head(100) # pris only les premiere 100 parce que autrement le map crashe ;-)

for _, airport in airports_df_petit_base.iterrows(): # on parcour tous les lignes dans le tableau et les rajoute au carte 

    f.Marker(
        location = [airport['lat'], airport['lon']], #the place of pin 
        popup=airport['name'], # names in the pop-up
        tooltip=airport['iata'], #hoover over 
    ).add_to(map_airports)

map_airports
```
