```python
import folium as f
```

```python
my_map = f.Map(
    location = [45.7640, 4.8357],
    zoom_start = 2
)
my_map 
```

```python
grandprixs = pd.read_csv("./archive/circuits.csv")
grandprixs.head(2)
```

```python
for _, gp in grandprixs.iterrows():

    f.Marker(
        location = [gp[5], gp[6]], #the place of pin 
        popup=gp['circuitRef'], # names in the pop-up
        tooltip=gp['circuitRef'], #hoover over 
    ).add_to(my_map)

my_map
```
