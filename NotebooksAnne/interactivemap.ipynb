```python
import folium as f
```

```python
my_map = f.Map(
    location = [45.7640, 4.8357],
    zoom_start = 2
)
my_map 
```

```python
grandprixs = pd.read_csv("./archive/circuits.csv")
grandprixs.head(2)
```

```python
for _, gp in grandprixs.iterrows():

    f.Marker(
        location = [gp[5], gp[6]], #the place of pin 
        popup=gp['circuitRef'], # names in the pop-up
        tooltip=gp['circuitRef'], #hoover over 
    ).add_to(my_map)

my_map
```

```python
seasons = pd.read_csv("./archive/races.csv")
seasons.head()

season21 = seasons[seasons['year'] == 2021]
season21
```

```python
# combining two datasets, looking for all the grandprix information of all the circuits and only keeping the ones in season 2021 
grandprix_21 = grandprixs[grandprixs['circuitId'].isin(season21['circuitId'])]
grandprix_21
```

```python
map_season21 = f.Map(
    location = [45.7640, 4.8357],
    zoom_start = 2
)

for _, gp in grandprix_21.iterrows():

    f.Marker(
        tiles ='Stamen Toner',
        location = [gp[5], gp[6]], #the place of pin 
        popup=gp['circuitRef'], # names in the pop-up
        tooltip=gp['circuitRef'], #hoover over 
    ).add_to(map_season21)

map_season21
```
